```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("/kaggle/input/playground-series-s4e3/train.csv")
train = train.set_index('id')
test = pd.read_csv("/kaggle/input/playground-series-s4e3/test.csv")
test = test.set_index('id')
train.head()
```

```python
from statsmodels.multivariate.manova import MANOVA
from statsmodels.formula.api import ols
import statsmodels.api as sm
import statsmodels.stats.multicomp as mc
import scipy.stats as stats

independent_cols = test.columns.tolist()
for c in independent_cols:
    min_value = train[c].min()
    max_value = train[c].max()
    train[c+"_scaled"] = (train[c] - min_value) / (max_value - min_value)
    test[c+"_scaled"] = (test[c] - min_value) / (max_value - min_value)
```

```python
train.columns
```

```python
independent_cols_scaled = test.columns.tolist()
independent_cols_scaled
```

```python
from statsmodels.multivariate.manova import MANOVA

# Assuming 'Response_Variable' is the dependent variable you are interested in
dependent_variable = 'Pastry+Z_Scratch+K_Scatch+Stains+Dirtiness+Bumps+Other_Faults'

# Creating the formula for MANOVA
formula = f'{dependent_variable} ~ {" + ".join(independent_cols_scaled)}'

# Creating the MANOVA object
manova = MANOVA.from_formula(formula, data=train)

# Printing the MANOVA results
print(manova.mv_test())
```

```python

    
    var_drop=(['Edges_Index', 
             'Outside_X_Index', 
             'Outside_Global_Index', 
             'Luminosity_Index', 
             'SigmoidOfAreas', 
             'X_Minimum_scaled', 
             'X_Maximum_scaled', 
             'Y_Minimum_scaled', 
             'Y_Perimeter_scaled', 
             'Sum_of_Luminosity_scaled', 
             'TypeOfSteel_A400_scaled', 
             'Edges_Index_scaled', 
             'Outside_X_Index_scaled', 
             'Outside_Global_Index_scaled', 
             'Luminosity_Index_scaled', 
             'SigmoidOfAreas_scaled',
             'Y_Minimum',
             'Y_Maximum', 
             'TypeOfSteel_A400',
             'TypeOfSteel_A300',
             'X_Perimeter',
             'X_Minimum',
             'X_Maximum',
             'Empty_Index',
             'Sum_of_Luminosity'])
def feats(df):
        df.drop(var_drop, axis=1, inplace=True)
        return df


train=feats(train)
test=feats(test)
to_remove = var_drop
independent_cols_scaled = [col for col in independent_cols_scaled if col not in to_remove]
```

```python
dependent_variables = ['Pastry','Z_Scratch','K_Scatch','Stains','Dirtiness','Bumps','Other_Faults']

for dv in dependent_variables:
    formula = f'{dv} ~ {" + ".join(independent_cols_scaled)}'

    data = ols(formula, data=train).fit()
    p_value = pd.DataFrame(sm.stats.anova_lm(data, typ=1))
    
    print(f"Dependent Variable: {dv}")
    print(p_value[p_value['PR(>F)'] < 0.05].sort_values(by='PR(>F)', ascending=True))
    print("\n")
```

```python
op_params = {'n_estimators': 50, 'learning_rate': 0.12002152645991093, 'max_depth': 7}

# Initialize XGBoost classifier with best parameters
best_classifier =GradientBoostingClassifier(**op_params, validation_fraction=0.1, n_iter_no_change=5, random_state=42)

# Create a MultiOutputClassifier
best_multi_output_classifier = MultiOutputClassifier(best_classifier, n_jobs=-1)

# Fit the model with the entire training data
best_multi_output_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_multi_output_classifier.predict(X_test)

# Evaluate the model's performance
model_score = accuracy_score(y_test, y_pred)
print(f"Model Score: {model_score}")
```
